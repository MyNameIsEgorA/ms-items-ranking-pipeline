```python
cols = [
    'Title', 'Vendor', 'Tags', 'Published',
    'Option1 Name', 'Option2 Name',
    'Gift Card',
    'code (product.metafields.custom.code)',
    'gender_list (product.metafields.custom.gender_list)',
    'type1 (product.metafields.custom.type1)',
    'type2 (product.metafields.custom.type2)',
    'type3 (product.metafields.custom.type3)',
    'Status'
]
filled_counts = df[cols].notna().sum(axis=1)
filled_counts.value_counts()
```

0 - вообще ни одна не заполнена
13 - заполнены все колонки
12 - заполнены почти все колонки
По этим колонкам, либо заполнено все, либо не заполнено ничего

```python
all_or_nothing = (filled_counts == 0) | (filled_counts == len(cols))

all_or_nothing.value_counts()
```

```python
df = df[df[cols].notna().sum(axis=1) > 0]
df.head()
```

```python
(df.notnull().mean() * 100).round(2)
```

```python
df = df.loc[:, df.nunique(dropna=False) > 1]
```
